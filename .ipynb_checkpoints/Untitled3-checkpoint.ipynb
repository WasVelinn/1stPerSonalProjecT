```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL 설정
url = 'https://www.gmarket.co.kr/n/search?keyword=%EB%A7%88%EC%9A%B0%EC%8A%A4'

# 웹페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 예시: 제품 이름, 가격, 리뷰 내용 추출
product_name = soup.find('span', {'class': 'product-title'}).text
product_price = soup.find('span', {'class': 'product-price'}).text
reviews = soup.find_all('div', {'class': 'review'})

# 리뷰 정보 추출
for review in reviews:
    review_text = review.find('p', {'class': 'review-text'}).text
    review_rating = review.find('span', {'class': 'rating'}).text
    print(f"Review: {review_text}, Rating: {review_rating}")
```

```python
import requests
import json

# 클라이언트 ID와 클라이언트 Secret (개인 API 정보)
client_id = "T3xJlimKqJrEwhcK6VTd"
client_secret = "uqGAxFqUDv"

# API 호출 URL (네이버 쇼핑 API)
url = "https://openapi.naver.com/v1/search/shop.json"

# 요청할 파라미터 설정 (검색어와 관련한 상품 정보)
params = {
    "query": "마우스",  # 예: "아이폰"
    "display": 10,  # 한 번에 받아올 상품 수
    "start": 1,  # 결과의 시작 번호
    "sort": "sim"  # 유사도 기준으로 정렬
}

# 헤더에 클라이언트 ID와 Secret 추가
headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}

# API 호출
response = requests.get(url, headers=headers, params=params)

# 응답 데이터 처리
data = response.json()

# 결과 출력 (상품 정보와 관련된 데이터를 출력)
for item in data["items"]:
    print(f"상품명: {item['title']}")
    print(f"가격: {item['lprice']}")
    print(f"링크: {item['link']}")
    print("---------------------------")
```
